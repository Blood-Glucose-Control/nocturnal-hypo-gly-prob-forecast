# Benchmarking Workflow for Nocturnal Probabilistic Forecasting

```python
from sktime.benchmarking.forecasting import ForecastingBenchmark
from sktime.datasets import load_airline
from sktime.forecasting.naive import NaiveForecaster
from sktime.performance_metrics.forecasting import MeanSquaredPercentageError, MeanSquaredError
from sktime.split import SingleWindowSplitter
```

```python
load_airline()
```

```python
benchmark = ForecastingBenchmark()
```

```python
benchmark.add_estimator(
    estimator=NaiveForecaster(strategy="mean", sp=12),
    estimator_id="NaiveForecaster-mean-v1",
)
benchmark.add_estimator(
    estimator=NaiveForecaster(strategy="last", sp=12),
    estimator_id="NaiveForecaster-last-v1",
)
```

## Benchmarking Tasks

### Cross-validation Split Regime

```python
cv_splitter = SingleWindowSplitter(
    fh=6*4
)
```

### Performance Metrics

```python
scorers = [MeanSquaredPercentageError(), MeanSquaredError(square_root=True)]
```

### Specify dataset loaders

```python
dataset_loaders = [load_airline]
```

### Add tasks to the benchmark instance

```python
for dataset_loader in dataset_loaders:
    benchmark.add_task(
        dataset_loader,
        cv_splitter,
        scorers,
    )
```

```python
results_df = benchmark.run("./forecasting_results.csv")
results_df['validation_id'][0]
```
