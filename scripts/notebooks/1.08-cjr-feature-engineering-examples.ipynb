```python
from src.data.diabetes_datasets.data_loader import get_loader

loader = get_loader(
    data_source_name="kaggle_brisT1D",
    dataset_type="train",
    keep_columns=[
        "datetime",
        "p_num",
        "bg_mM",
        "hr_bpm",
        "steps",
        "cals",
        "cob",
        "carb_availability",
        "dose_units",
        "iob",
        "insulin_availability",
    ],
    use_cached=False,
)
```

```python
import pandas as pd

raw_data = pd.read_csv("/u6/cjrisi/nocturnal/cache/data/kaggle_brisT1D/raw/train.csv")
raw_data.head()
```

```python
keep_columns = raw_data.columns
for col in keep_columns:
    print(col)
```

```python
keep_columns = [
    "id",
    "p_num",
    "time",
    "bg-0:00",
    "insulin-0:00",
    "carbs-0:00",
    "hr-0:00",
    "steps-0:00",
    "cals-0:00",
    "activity-0:00",
]
raw_data = raw_data[keep_columns]
```

```python
raw_data.head()
```

```python
# Fill NaNs with 0s for calculation purposes
data_filled = raw_data.fillna(0)

# Group by patient and calculate summary statistics
patient_stats = data_filled.groupby("p_num").agg(
    {
        "bg-0:00": ["sum", "mean", "count"],
        "insulin-0:00": ["sum", "mean", "count"],
        "carbs-0:00": ["sum", "mean", "count"],
        "hr-0:00": ["sum", "mean", "count"],
        "steps-0:00": ["sum", "mean", "count"],
        "cals-0:00": ["sum", "mean", "count"],
    }
)

# Flatten column names for easier access
patient_stats.columns = ["_".join(col).strip() for col in patient_stats.columns.values]

# Display the results
print("Summary Statistics by Patient:")
print(patient_stats.round(2))

# Optional: Save to a more readable format
patient_summary = patient_stats.copy()
print("\nFirst 5 patients:")
print(patient_summary.head())
```

```python
patient_summary.round(3).to_csv(
    "/u6/cjrisi/nocturnal/cache/data/kaggle_brisT1D/processed/raw_patient_summary.csv",
    index=True,
)
```
